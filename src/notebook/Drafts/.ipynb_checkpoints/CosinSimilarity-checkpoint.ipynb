```python
from nltk.tokenize import word_tokenize

def newEmbeddings(pathfile):
    with open(pathfile,errors='ignore') as doc:
    #dimension of the terms & nb terms 
        dimension=[]*2
        line=doc.readline()
        dimension=word_tokenize(line)
        lines = doc.readlines()
        l=0
        linembeddings=[""]*int(dimension[0])
        print(len(linembeddings))
        for line in lines:
            linembeddings[l]=line  
            l=l+1
            
        newembeddings=[['']]*len(linembeddings)
        i=0
        
        for str in linembeddings :
            newembeddings[i]=word_tokenize(str)
           # print(newembeddings[i])
            newembeddings[i][0]=newembeddings[i][0].lower()
            i=i+1
            #pour voir ou en est le code
            #print(i)
            #pour tester sinon trop long
            #if(i==3): break
            
    return newembeddings
#Faut il proceder à la lemmization ?
```

```python
from scipy.io.matlab import mio
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#Récupérer le nombre d'abstract pour les 5 documents
label = mio.loadmat("../../../../data/PubMed10/label.mat")
# Define the classes to be selected
# For PubMed5: 
#  (6) Hay Fever --> nDoc = 1517
#  (7) Kidney Calculi --> nDoc = 1549
#  (8) Age-related Macular Degeneration --> nDoc = 3283
#  (9) Migraine --> nDoc = 3704
# (10) Otitis --> nDoc = 2597
abstractDocHayFever = label.get("size6")[0][0]
abstractDocKidneyCalculi = label.get("size7")[0][0]
abstractDocAMD = label.get("size8")[0][0]
abstractDocMigraine = label.get("size9")[0][0] + 1 #Le label est 3703 alors qu'en vrai c'est 3704
abstractDocOtitis = label.get("size10")[0][0] + 1

print(abstractDocHayFever, abstractDocKidneyCalculi, abstractDocAMD, abstractDocMigraine, abstractDocOtitis)
```

```python
#Devide Document to abstracts
# import these modules 
#from nltk.stem import PorterStemmer 
from nltk.stem import WordNetLemmatizer 
  
def corpus(pathfile,numberAbs):
    with open(pathfile,errors='ignore') as doc:
        lines = doc.readlines()
    # Itérer sur les lignes
    #il faudrait créer un vecteur de documents
    
    i=0
    listcorpus=[""]*numberAbs
    for line in lines:
        if (line.find('PMID') == -1):
            listcorpus[i]= listcorpus[i] +line + " "
        else:
            #listChickpox[i]= listChickpox[i] + line + " "
            i+= 1
            
    return listcorpus

#tockenize each abstract and clean it 
def cleanAbstracts(listAbs):    
    
    listcorpus=[[]]*len(listAbs)
    stop = set(stopwords.words('english'))
    stop.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','%','#','-','_']) 
    for i in range(len(listAbs)):
        tokens = word_tokenize(listAbs[i])
        tokens = [w for w in tokens if not w in stop]
        listcorpus[i]=tokens
    #Stemming of words [Voc_ext]->[Voc] to match the words from the output of the co-clust
    #ps =PorterStemmer()
    #lemmatizer = WordNetLemmatizer()
    for i in range(len(listAbs)):
       # listcorpus[i] = [ps.stem(w) for w in listcorpus[i]]
       # listcorpus[i] = [lemmatizer.lemmatize(w) for w in listcorpus[i]]
        listcorpus[i]=[w.lower() for w in listcorpus[i]]

    return listcorpus
```

```python
#### Build a corpus as a list of list to use as an input for Word2Vec
listabsHayFever=corpus("../../../../data/PubMed10/pubmed_result_hayfever.txt",abstractDocHayFever)
corpusHayFever=cleanAbstracts(listabsHayFever)

listabsKidneyCalculi=corpus("../../../../data/PubMed10/pubmed_result_kidneycalculi.txt",abstractDocKidneyCalculi)
corpusKidneyCalculi=cleanAbstracts(listabsKidneyCalculi)

listabsAMD=corpus("../../../../data/PubMed10/pubmed_result_macular.txt",abstractDocAMD)
corpusAMD=cleanAbstracts(listabsAMD)

listabsMigraine=corpus("../../../../data/PubMed10/pubmed_result_migraine.txt",abstractDocMigraine)
corpusMigraine=cleanAbstracts(listabsMigraine)

listabsOtitis=corpus("../../../../data/PubMed10/pubmed_result_otitis.txt",abstractDocOtitis)
corpusOtitis=cleanAbstracts(listabsOtitis)

corpusPubMed5=corpusHayFever+corpusKidneyCalculi+corpusMigraine+corpusOtitis+corpusHayFever
```

```python
from gensim.models import Word2Vec

model = Word2Vec(corpusPubMed5, size=300, window=20, min_count=1, workers=4)
n_top = 20
words = list(sorted(model.wv.vocab))
#Get the n_top most frequent words
topterms = model.wv.index2entity[:n_top]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def computeCosine2Vec(v,w):
    a = np.array(v)
    b = np.array(w)
    aa = a.reshape(1,len(v))
    ba = b.reshape(1,len(v))
    cosim = cosine_similarity(aa, ba)
    return cosim[0][0]
```

```python
def getMatchingEmbedding(w,embeddings):
    vect=['']
    j=0
    for emb in embeddings:
        if(emb[0]==w)and(j==0):
            #print(w)
            vect=[emb[i] for i in range(1,len(emb)) ]
            j=j+1
            #print(vect[0])
        if not (j==0): break
    return vect
```

```python

def computeCosineSimilarity(pathfile,vocab,embeddings):
    cr = csv.reader(open(pathfile,"r"))
    for row in cr:
        topterms=row
        break
    newtopterms=['']*len(topterms)
    i=0
    for w in topterms:
        if(w in vocab):
            newtopterms[i]=w
            i=i+1
        if(w not in vocab):
            for v in vocab:
                if (w in v)and(w[0]==v[0]):
                    newtopterms[i]=v
                    i=i+1
                    break
        
            

    #probleme avec significantli moi je dis on le vire :3
    #Compute the cosin similarity matrix     
    rows = []
    for w in newtopterms:
        if (w==''):
            #print('yes')
            newtopterms.remove(w)
            
    if ('' in newtopterms):
        newtopterms.remove('')
    if ('' in newtopterms):
        newtopterms.remove('')
   
    print(newtopterms)
    for w in newtopterms:
        vect_w = getMatchingEmbedding(w,embeddings)
        for ww in newtopterms:
            if(w == ww): rows.append(dict(w1=w, w2=ww, d= 0))
            else :
                vect_ww = getMatchingEmbedding(ww,embeddings)
                cos=computeCosine2Vec(vect_w,vect_ww)
                rows.append(dict(w1=w, w2=ww, d=cos))
     
    cosindf = pd.DataFrame(rows).pivot("w1", "w2", "d")
    return cosindf
```

Graph

```python
import igraph
import numpy as np
#il faudra crée un fichier avec la frequence des tops termes pour les noeuds 
def graphCos(cosindf,pathfile, color):
    #Topterms 
    vect=cosindf.index
    topterms=['']*len(vect)
    for i in range(len(vect)):
        topterms[i]=vect[i]
    
    #frequences
    frequences=[0]*len(topterms)
    i=0
    with open(pathfile) as csvDataFile:
        csvReader = csv.reader(csvDataFile)
        for row in csvReader:
            if not(len(row)==0):
                #print(row[0])
                for v in topterms :
                    if(row[0] in v)and(row[0][0]==v[0]):
                        frequences[i]=float(row[1])
                        #print(frequences[i])
                        i=i+1
                        break

    # Make a graph from the adjacency matrix
    g = igraph.Graph.Weighted_Adjacency((cosindf.values > 0.3).tolist(), mode = "undirected")

    # Set words as vertices label
    g.vs["name"] = topterms
    g.vs["label"] = g.vs["name"]
    g.vs["color"] = color

    # Set size as vertices size by frequence of term in documents
    size_vertex =frequences 
    #normalize the vectors
    size_vertex_norm = [float(i)/(sum(size_vertex) - size_vertex[0])*800  for i in size_vertex]
    g.vs["size"] = size_vertex_norm

    # Set cosine similarity as edge width
    edge_w = [0]*len(g.get_edgelist())
    
    for ie, (i,j) in enumerate(g.get_edgelist()):
        if i == j:
            edge_w[ie] = 0
        else:
            edge_w[ie] = (cosindf.iloc[i][j] + 2)
    edge_w_color = [""]*len(g.get_edgelist())
    #il faut trouver un moyen de trouver les intervalles automatiquement
    for ie, edge_value in enumerate(edge_w):
        if (edge_value > 2.4):
            edge_w_color[ie] = "#126600" 
        elif (edge_value > 2.35 and edge_value < 2.4):
            edge_w_color[ie] = "#25cc00"
        else:
            edge_w_color[ie] = "#2eff00"

    g.es["weight"] = edge_w
    g.es["width"] = g.es["weight"]
    g.es["color"] = edge_w_color

    return g
    
    #store the graph in a file.gml
    #igraph.Graph.write_gml(g,"testgraph_word2Vec.gml")
```
