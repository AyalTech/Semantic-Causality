```python
from scipy.io.matlab import mio
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#Récupérer le nombre d'abstract pour les 5 documents
label = mio.loadmat("../../../../data/PubMed10/label.mat")
# Define the classes to be selected
# For PubMed5: 
#  (6) Hay Fever --> nDoc = 1517
#  (7) Kidney Calculi --> nDoc = 1549
#  (8) Age-related Macular Degeneration --> nDoc = 3283
#  (9) Migraine --> nDoc = 3704
# (10) Otitis --> nDoc = 2597
abstractDocHayFever = label.get("size6")[0][0]
abstractDocKidneyCalculi = label.get("size7")[0][0]
abstractDocAMD = label.get("size8")[0][0]
abstractDocMigraine = label.get("size9")[0][0] + 1 #Le label est 3703 alors qu'en vrai c'est 3704
abstractDocOtitis = label.get("size10")[0][0] + 1

print(abstractDocHayFever, abstractDocKidneyCalculi, abstractDocAMD, abstractDocMigraine, abstractDocOtitis)
```

```python
#Devide Document to abstracts
# import these modules 
#from nltk.stem import PorterStemmer 
from nltk.stem import WordNetLemmatizer 
  
def corpus(pathfile,numberAbs):
    with open(pathfile,errors='ignore') as doc:
        lines = doc.readlines()
    # Itérer sur les lignes
    #il faudrait créer un vecteur de documents
    
    i=0
    listcorpus=[""]*numberAbs
    for line in lines:
        if (line.find('PMID') == -1):
            listcorpus[i]= listcorpus[i] +line + " "
        else:
            #listChickpox[i]= listChickpox[i] + line + " "
            i+= 1
            
    return listcorpus

#tockenize each abstract and clean it 
def cleanAbstracts(listAbs):    
    
    listcorpus=[[]]*len(listAbs)
    stop = set(stopwords.words('english'))
    stop.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','%','#','-','_']) 
    for i in range(len(listAbs)):
        tokens = word_tokenize(listAbs[i])
        tokens = [w for w in tokens if not w in stop]
        listcorpus[i]=tokens
    #Stemming of words [Voc_ext]->[Voc] to match the words from the output of the co-clust
    #ps =PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    for i in range(len(listAbs)):
       # listcorpus[i] = [ps.stem(w) for w in listcorpus[i]]
        listcorpus[i] = [lemmatizer.lemmatize(w) for w in listcorpus[i]]
    return listcorpus
```

```python
#### Build a corpus as a list of list to use as an input for Word2Vec
listabsHayFever=corpus("../../../../data/PubMed10/pubmed_result_hayfever.txt",abstractDocHayFever)
corpusHayFever=cleanAbstracts(listabsHayFever)

listabsKidneyCalculi=corpus("../../../../data/PubMed10/pubmed_result_kidneycalculi.txt",abstractDocKidneyCalculi)
corpusKidneyCalculi=cleanAbstracts(listabsKidneyCalculi)

listabsAMD=corpus("../../../../data/PubMed10/pubmed_result_macular.txt",abstractDocAMD)
corpusAMD=cleanAbstracts(listabsAMD)

listabsMigraine=corpus("../../../../data/PubMed10/pubmed_result_migraine.txt",abstractDocMigraine)
corpusMigraine=cleanAbstracts(listabsMigraine)

listabsOtitis=corpus("../../../../data/PubMed10/pubmed_result_otitis.txt",abstractDocOtitis)
corpusOtitis=cleanAbstracts(listabsOtitis)

corpusPubMed5=corpusHayFever+corpusKidneyCalculi+corpusMigraine+corpusOtitis+corpusHayFever
```

```python
#On ne peut utiliser les n terms du clusters psq c'est des mots coupés qu'on ne retrouve pas dans notre corpus de mots
#Example otiti est dans les n terms du cluster Otitis mais ce n'est pas un mots qu'on retrouve dans notre corpus
#Donc on ne peut pas calculer les cos similarity avec W2V
#Ainsi on utilise un algorithme pour transformer les vocab de model.wv.vocab en vocab de topterms
newtopterms=['']*len(topterms)
i=0
vocab=list(model.wv.vocab)
vocab=[w.lower() for w in vocab]
for w in topterms:
    if (w not in vocab):
         for v in vocab:
            if (w in v)and (w[0]==v[0]) and (w[1]==v[1]):
                newtopterms[i]=v
                break
        
    else:
        newtopterms[i]=w

       
        
        
    i=i+1
        
print(newtopterms)   
```

```python
rows = []

for w in newtopterms:
    for ww in newtopterms:
        rows.append(dict(w1=w, w2=ww, d=model.wv.similarity(w, ww)))
#Compute the cosin similarity matrix 
cosinPubMed5 = pd.DataFrame(rows).pivot("w1", "w2", "d")
```

```python
import igraph
import numpy as np

# Make a graph from the adjacency matrix
g = igraph.Graph.Weighted_Adjacency((cosinPubMed5.values > 0.3).tolist(), mode = "undirected")

# Set words as vertices label
g.vs["name"] = topterms
g.vs["label"] = g.vs["name"]
g.vs["color"] = "pink"

# Set size as vertices size by frequence of term in documents
size_vertex = [0] * n_top
i=0
for item in newtopterms:
    size_vertex[i]= model.wv.vocab[item].count
    i=i+1
#normalize the vectors
size_vertex_norm = [float(i)/(sum(size_vertex) - size_vertex[0])*800  for i in size_vertex]
g.vs["size"] = size_vertex_norm

# Set cosine similarity as edge width
edge_w = [0]*len(g.get_edgelist())
for ie, (i,j) in enumerate(g.get_edgelist()):
    if i == j:
        edge_w[ie] = 0
    else:
        edge_w[ie] = (cosinPubMed5.iloc[i][j])
edge_w_color = [""]*len(g.get_edgelist())
for ie, edge_value in enumerate(edge_w):
    if (edge_value > 0.4):
        edge_w_color[ie] = "#b50000"
    elif (edge_value > 0.35 and edge_value < 0.4):
        edge_w_color[ie] = "#000cff"
    else:
        edge_w_color[ie] = "#00ff61"

g.es["weight"] = edge_w
g.es["width"] = g.es["weight"]
g.es["color"] = edge_w_color

# plot the graph
igraph.summary(g)
igraph.plot(g, margin= 50, bbox = (700,700))
#igraph.Graph.write_gml(g,"testgraph_word2Vec.gml")
```
